Libraries for Handling Tabular Data

*   NumPy : Library for Numeric Computations in Python
*   Pandas : Library for Data Acquisition and Preparation
*   Matplotlib : Basic Library for Data Visualization
*   Seaborn: Advanced Library for Data Visualization



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()  # set the default Seaborn style for graphics
```

Import the Dataset

To read the data in comma-separated values (CSV) format, we use the read_csv function from Pandas. Then, we can take a quick look at the data using the head() function.

```python
# URL for the Iris dataset from UCI Machine Learning Repository
path = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

# Column names for the Iris dataset
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Load the dataset into a pandas DataFrame
df = pd.read_csv(path, names=column_names)

# Display the first 5 rows of the DataFrame
df.head()
```

```python
# Print the type and shape attributes
print("Data type : ", type(df))
print("Data dims : ", df.shape)
```

```python
# Print the type of dataframe
print(df.dtypes)
```

```python
# Print the statistics about the dataset
df.describe()
```

```python
# Extract a particular column from the dataframe
S_col = pd.DataFrame(df['species'])
print("Data type : ", type(S_col))
print("Data dims : ", S_col.size)
S_col.head()
```

Visualizing Variable

Using Hisplot and Boxplot

```python
sns.histplot(data=df, kde=True)
```

```python
# Extract only the numeric data variables
numDF = pd.DataFrame(df[["sepal_length", "sepal_width", "petal_length", "petal_width"]])

# Summary statistics for all variables
numDF.describe()
```

```python
sns.boxplot(data = numDF)
```

```python
# Visualize correlation between numerical values
sns.heatmap(numDF.corr(), annot=True, cmap="Blues")
```

```python
sns.pairplot(df, hue="species")
```
